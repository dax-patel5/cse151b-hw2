```python
import torch

from generate import generate_text
from util import encode_text
from config import load_config

from shakespeare_lstm import LSTMModel
from shakespeare_rnn import RNNModel
```

```python
input_file = 'data/tiny_shakespeare.txt'
device = torch.device("cuda" if torch.cuda.is_available()
                      else "mps" if torch.backends.mps.is_available() else "cpu")

encoded_text, vocab_size, char_to_idx, idx_to_char = encode_text(input_file)

### Load the right config
config_path = 'configs/base_lstm_config.yaml'
config = load_config(config_path)


#### Load your trained model (best performing LSTM)
model = LSTMModel(vocab_size=vocab_size,
                 embed_size=config['embed_size'],
                hidden_size=config['hidden_size'],
                num_layers=config['num_layers'])

model_path = 'models/trained_lstm_model.pth'
model.load_state_dict(torch.load(model_path, weights_only=True))

# Example usage
generated = generate_text(model, device, char_to_idx, idx_to_char, max_len=1000, temp=0.8)
print(generated)
```

```python
# Question 5(d): generate ~1000-character samples from the best performing LSTM
# at temperatures T = 1, 2, and 0.5, and save them to text files.
for temp in (0.5, 1, 2):
    text = generate_text(model, device, char_to_idx, idx_to_char, max_len=1000, temp=temp)
    fname = f"temp{temp}.txt"
    with open(fname, 'w') as f:
        f.write(text)
    print(f"===== T = {temp} (saved to {fname}) =====")
    print(text[:500])
    print()
```
